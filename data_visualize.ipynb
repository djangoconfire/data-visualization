```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic=sns.load_dataset('titanic')
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

### Categorical Scatterplot

#### A simple way to show some quantitative values across the levels of categorical variables use stripplot..here one of the variable is categorical 

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
## removing overlapping of scatter plot ..use jitter=True
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
```

```python
# A different approach to avoid overlapping :  use swarmplot
sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
## Adding a nested categorical variable with hue parameter.
sns.swarmplot(x="day",y="total_bill",hue='sex',data=tips)
```

```python
# Always put categorical variable on vertical axis
sns.swarmplot(x="total_bill",y="day",hue='sex',data=tips)
```

```python
## Box Plot
sns.boxplot(x='day',y='total_bill',hue='time',data=tips)
```

```python
# violin plot
sns.violinplot(x='total_bill',y='day',hue='time',data=tips)
```

```python
# it is possible to spplit the violin when hue parameter has only two levels
sns.violinplot(x='total_bill',y='day',hue='sex',data=tips,split=True)
```

### Statistical estimation within Categories

```python
# Bar plot :- It operates on full dataset and shows an arbitary estimate using mean by default
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
# showing number of observation in each category rather than computing statistics for second variable
# similar to histogram for categorical variable 
sns.countplot(x='deck',data=titanic,palette="Greens_d")
```

```python
sns.countplot(y='deck',hue='class',data=titanic,palette='Greens_d')
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',markers=["o","x"],col='time',row='sex',data=tips,palette="Set1")
```
